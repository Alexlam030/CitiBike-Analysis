```python
#Testing for mixed data types
for col in df_citi_bike_drop.columns.tolist():
  weird = (df_citi_bike_drop[[col]].applymap(type) != df_citi_bike_drop[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_citi_bike_drop[weird]) > 0:
    print (col)
```

```python
#Changing weekday column name for clarity
df_citi_bike_drop.rename(columns = {'weekday' : 'day_of_week'}, inplace = True)
```

```python
#Checking column name changes
df_citi_bike_drop.columns
```

```python
#Checking what days of week is in dataset
df_citi_bike_drop['day_of_week'].unique()
```

```python
#Changing gender column to string
df_citi_bike_drop['gender'] = df_citi_bike_drop['gender'].astype('str')
```

```python
#Changing gender number values to N/A, M, F for clarity
df_citi_bike_drop['gender'] = df_citi_bike_drop['gender'].replace(['0', '1', '2'],['N/A', 'M', 'F'])
```
